```python
def age_null(cols):
    Age=cols[0];
    Pclass=cols[1];
    if(pd.isnull(Age)):
        if Pclass == 1:
            return 38
        if Pclass == 2:
            return 29
        if Pclass ==3:
            return 24
    else:
        return Age
```

```python
df['Age']=df[['Age','Pclass']].apply(age_null,axis=1)
test['Age']=test[['Age','Pclass']].apply(age_null,axis=1)
```

```python
sex_new=pd.get_dummies(df['Embarked'],drop_first=True)
embark_new=pd.get_dummies(df['Sex'],drop_first=True)
```

```python
df.drop(['Name','Sex','Ticket','Embarked'],axis=1,inplace=True)
```

```python
df.head()
```

```python
train=pd.concat([df,sex_new,embark_new],axis=1)
```

```python
train.head()
```

```python
sex_new_test=pd.get_dummies(test['Embarked'],drop_first=True)
embark_new_test=pd.get_dummies(test['Sex'],drop_first=True)
test.drop(['Name','Sex','Ticket','Embarked'],axis=1,inplace=True)
test.head()
```

```python
test=pd.concat([test,sex_new_test,embark_new_test],axis=1)
test
```

```python
test['Fare'] = test['Fare'].fillna(test['Fare'].median())
```

```python
train_x=train.drop(['Survived'],axis=1)
```

```python
train_x.head()
```

```python
train_y=train['Survived']
```

```python
train_y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_x, 
                                                    train_y, test_size=0.2, 
                                                    random_state=10)
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
model=RandomForestClassifier(criterion='gini', 
                             n_estimators=700,
                             min_samples_split=10,
                             min_samples_leaf=1,
                             max_features='auto',
                             oob_score=True,
                             random_state=1,
                             n_jobs=-1)
```

```python
model.fit(X_train,y_train)
model
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,model.predict(X_test))
```

```python
output = pd.DataFrame()
output['PassengerId'] = test['PassengerId']
output['Survived'] = model.predict(test)
```
